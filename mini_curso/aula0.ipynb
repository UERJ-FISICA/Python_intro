# Sistemas Operacionais

## O que faz o sistema operacional

O sistema operacional (SO) do seu computador **gerencia** todo o *software* e *hardware* do computador. Na maioria das vezes, há vários programas de computador diferentes sendo executados ao mesmo tempo e todos precisam acessar a unidade de processamento central (CPU), a memória e o armazenamento do computador. O sistema operacional coordena tudo isso para garantir que cada programa receba o que precisa.

## Tipos de sistemas operacionais

Os sistemas operacionais geralmente vêm pré-carregados em qualquer computador que você compre. A maioria das pessoas usa o sistema operacional que vem com o computador, mas é possível atualizar ou até alterar os sistemas operacionais. Os três sistemas operacionais mais comuns para computadores pessoais são **Microsoft Windows**, **macOS** e **Linux**.

**Os sistemas operacionais modernos usam uma interface gráfica do usuário ou GUI** (pronunciada guí). Uma GUI permite usar o mouse para clicar em ícones, botões e menus, e tudo é exibido claramente na tela usando uma combinação de gráficos e texto.

**Cada sistema operacional tem uma GUI diferente.** A GUI de cada sistema operacional tem uma aparência e comportamento diferentes, portanto, se você mudar para um sistema operacional diferente, pode parecer pouco familiar no início. No entanto, os sistemas operacionais modernos são projetados para serem fáceis de usar e a maioria dos princípios básicos é a mesma.

## Linux

O Linux é uma família de sistemas operacionais de **código aberto**, o que significa que eles podem ser modificados e distribuídos por qualquer pessoa em todo o mundo. Isso **é diferente do software proprietário** como o Windows, que só pode ser modificado pela empresa que o possui. As vantagens do Linux são que ele é gratuito e há muitas distribuições diferentes - ou versões - que você pode escolher.

Para saber mais sobre as diferentes distribuições do Linux, visite os sites do [Ubuntu](https://ubuntu.com/), [Linux Mint](https://www.linuxmint.com/) e [Fedora](https://getfedora.org/), por exemplo, mas existem muitas outras distribuições.

Aqui no laboratório do IF da UERJ, usamos o [light Ubuntu](https://lubuntu.net/).

## Comandos em Linux

Embora todos os sistemas operacionais tenham uma interface gráfica para navegarmos entre os diretórios e arquivos, é bastante útil sabermos operar os terminais de comandos, criar pastas, abrir arquivos e rodar programas através da linha de comando.

Como primeiro exercício, vamos:
1. Abrir um **terminal**;
1. Criar uma pasta chamada **python_escolas**;
1. Dentro da pasta **python_escolas**, criar uma nova pasta com o seu nome. 

Para tal, é útil saber os seguintes comandos de Linux, uma vez com o **terminal** aberto.

* *cd* : acessa a pasta principal do usuário, que chamamos de *home* (*casa*);
* *cd < pasta > * : acessar a pasta que se encontra no *caminho* pasta (os delimitadores '<' e '>' são somente para indicar o caminho a ser acessado);
* *ls*: lista o conteúdo da pasta atual;
* *ls < pasta >*: lista o conteúdo da pasta especificada;
* *mkdir < nome_pasta >*: cria uma pasta chamada *nome_pasta* dentro da pasta atual;
* *pwd*: lista o caminho da pasta atual


Uma vez com a sua pasta criada, podem digitar *python3* e ver o que acontece. :-)
