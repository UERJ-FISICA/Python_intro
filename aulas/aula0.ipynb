# Introdução a Python

Esse é um curso de **introdução** a Python, aberto a todas as pessoas que queiram aprender essa linguagem de programação. Não é necessário nenhum conhecimento prévio de Python ou qualquer outra linguagem de programação.

É desejável que a aluna ou o aluno tenham algum conhecimento de Linux, pois utilizaremos computadores com esse sistema operacional, porém não é mandatório. Aos que não têm conhecimento prévio de linux, recomendamos que pesquisem os comandos simples, como listar arquivo em um diretório ou criar um diretório.

A versão de Python utilizada no curso será **Python3**.
